```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from scipy import stats

# ÎπÑÍµêÌï† ÌöåÍ∑ÄÍ≥ÑÏàò Î™©Î°ù
selected_variables = [
    "ln_ÏãúÍ∞ÄÏ¥ùÏï°",
    "Ï¥ùÏï°ÎåÄÎπÑÏòÅÏóÖÏù¥Ïùµ",
    "Ï¥ùÏûêÏÇ∞Î≥ÄÌôîÏú®",
    "ÏòÅÏóÖÏù¥ÏùµÎ•†ÌëúÏ§ÄÌé∏Ï∞®",
    "Ïù¥Ìöå",
    "ÏÇ¨Ìöå",
    "PER",
    "Î∂ÄÏ±ÑÎπÑÏú®",
    "Í≥µÏãúÏó¨Î∂Ä",
]

# ÌöåÍ∑Ä Î∂ÑÏÑù Ïã§Ìñâ
y_kospi = df_KOSPI["log_Ï¥ùÎ∞∞ÎãπÍ∏àÏï°"]
X_kospi = sm.add_constant(df_KOSPI[selected_variables])
model_kospi = sm.OLS(y_kospi, X_kospi).fit()

y_kosdaq = df_KOSDAQ["log_Ï¥ùÎ∞∞ÎãπÍ∏àÏï°"]
X_kosdaq = sm.add_constant(df_KOSDAQ[selected_variables])
model_kosdaq = sm.OLS(y_kosdaq, X_kosdaq).fit()

# ÌöåÍ∑Ä Í≥ÑÏàò Î∞è ÌëúÏ§Ä Ïò§Ï∞® Í∞ÄÏ†∏Ïò§Í∏∞
params_kospi = model_kospi.params
params_kosdaq = model_kosdaq.params
bse_kospi = model_kospi.bse
bse_kosdaq = model_kosdaq.bse

# ÏûêÏú†ÎèÑ Í≥ÑÏÇ∞
df_kospi = len(y_kospi) - len(selected_variables) - 1  # ÏûêÏú†ÎèÑ Í≥ÑÏÇ∞
df_kosdaq = len(y_kosdaq) - len(selected_variables) - 1
df_total = df_kospi + df_kosdaq  # Ï¥ù ÏûêÏú†ÎèÑ


# ÌöåÍ∑Ä Í≥ÑÏàò Ï∞®Ïù¥ Í≤ÄÏ†ï ÏàòÌñâ
def compare_selected_coefficients(
    params1, params2, bse1, bse2, selected_vars, label1, label2, df
):
    params1_selected = params1[selected_vars]
    params2_selected = params2[selected_vars]
    bse1_selected = bse1[selected_vars]
    bse2_selected = bse2[selected_vars]

    # Í≥ÑÏàò Ï∞®Ïù¥ Î∞è ÌëúÏ§Ä Ïò§Ï∞® Í≥ÑÏÇ∞
    diff = params1_selected - params2_selected
    std_err_diff = np.sqrt(bse1_selected**2 + bse2_selected**2)

    # t-Í≤ÄÏ†ï ÏàòÌñâ
    t_stat = diff / std_err_diff
    p_values = stats.t.sf(abs(t_stat), df=df) * 2  # ÏñëÏ∏° Í≤ÄÏ†ï

    # Í≤∞Í≥º Ï†ÄÏû•
    results = pd.DataFrame(
        {
            "Coefficient Difference": diff,
            "Standard Error": std_err_diff,
            "t-statistic": t_stat,
            "p-value": p_values,
        }
    )

    # Í≤∞Í≥º Ï∂úÎ†•
    print(f"==== {label1} vs {label2} ÌöåÍ∑ÄÍ≥ÑÏàò Ï∞®Ïù¥ Í≤ÄÏ†ï ====")
    print(results.round(4))
    print("\n" + "=" * 50 + "\n")

    # Í≤∞Í≥º CSVÎ°ú Ï†ÄÏû• (MacÏóêÏÑú Î≥¥Í∏∞ ÏâΩÍ≤å)
    results.to_csv("coefficient_comparison_results.csv", encoding="utf-8-sig")
    print("üìå Í≤∞Í≥ºÍ∞Ä coefficient_comparison_results.csv ÌååÏùºÎ°ú Ï†ÄÏû•ÎêòÏóàÏäµÎãàÎã§!")

    return results


# KOSPIÏôÄ KOSDAQÏùò ÌäπÏ†ï Í≥ÑÏàò Ï∞®Ïù¥ Í≤ÄÏ†ï Ïã§Ìñâ
wald_test_results_df = compare_selected_coefficients(
    params_kospi,
    params_kosdaq,
    bse_kospi,
    bse_kosdaq,
    selected_variables,
    "KOSPI",
    "KOSDAQ",
    df_total,
)
```
