```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from scipy import stats

# 비교할 회귀계수 목록
selected_variables = [
    "ln_시가총액",
    "총액대비영업이익",
    "총자산변화율",
    "영업이익률표준편차",
    "이회",
    "사회",
    "PER",
    "부채비율",
    "공시여부",
]

# 회귀 분석 실행
y_kospi = df_KOSPI["log_총배당금액"]
X_kospi = sm.add_constant(df_KOSPI[selected_variables])
model_kospi = sm.OLS(y_kospi, X_kospi).fit()

y_kosdaq = df_KOSDAQ["log_총배당금액"]
X_kosdaq = sm.add_constant(df_KOSDAQ[selected_variables])
model_kosdaq = sm.OLS(y_kosdaq, X_kosdaq).fit()

# 회귀 계수 및 표준 오차 가져오기
params_kospi = model_kospi.params
params_kosdaq = model_kosdaq.params
bse_kospi = model_kospi.bse
bse_kosdaq = model_kosdaq.bse

# 자유도 계산
df_kospi = len(y_kospi) - len(selected_variables) - 1  # 자유도 계산
df_kosdaq = len(y_kosdaq) - len(selected_variables) - 1
df_total = df_kospi + df_kosdaq  # 총 자유도


# 회귀 계수 차이 검정 수행
def compare_selected_coefficients(
    params1, params2, bse1, bse2, selected_vars, label1, label2, df
):
    params1_selected = params1[selected_vars]
    params2_selected = params2[selected_vars]
    bse1_selected = bse1[selected_vars]
    bse2_selected = bse2[selected_vars]

    # 계수 차이 및 표준 오차 계산
    diff = params1_selected - params2_selected
    std_err_diff = np.sqrt(bse1_selected**2 + bse2_selected**2)

    # t-검정 수행
    t_stat = diff / std_err_diff
    p_values = stats.t.sf(abs(t_stat), df=df) * 2  # 양측 검정

    # 결과 저장
    results = pd.DataFrame(
        {
            "Coefficient Difference": diff,
            "Standard Error": std_err_diff,
            "t-statistic": t_stat,
            "p-value": p_values,
        }
    )

    # 결과 출력
    print(f"==== {label1} vs {label2} 회귀계수 차이 검정 ====")
    print(results.round(4))
    print("\n" + "=" * 50 + "\n")

    # 결과 CSV로 저장 (Mac에서 보기 쉽게)
    results.to_csv("coefficient_comparison_results.csv", encoding="utf-8-sig")
    print("📌 결과가 coefficient_comparison_results.csv 파일로 저장되었습니다!")

    return results


# KOSPI와 KOSDAQ의 특정 계수 차이 검정 실행
wald_test_results_df = compare_selected_coefficients(
    params_kospi,
    params_kosdaq,
    bse_kospi,
    bse_kosdaq,
    selected_variables,
    "KOSPI",
    "KOSDAQ",
    df_total,
)
```
