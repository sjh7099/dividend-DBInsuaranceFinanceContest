```python
import OpenDartReader
import pandas as pd

# API 인증키 설정
api_key = "1ec628ca0ac283731a99b879ea1ab65f79897d24"
dart = OpenDartReader(api_key)

# 데이터 수집 연도 설정 (과거 데이터 조회)
start_year = 2015
end_year = 2022
corp_code = "005930"  # 삼성전자 예시

# 내부지분율 데이터 저장 리스트
results = []

for year in range(start_year, end_year + 1):
    print(f"📌 {year}년 {corp_code} 과거 데이터 수집 중...")

    try:
        # 📌 해당 연도의 공시 목록 가져오기 (사업보고서 기준)
        reports = dart.list(corp_code)  # 📌 전체 공시 목록을 가져옴
        reports["연도"] = (
            reports["rcept_no"].astype(str).str[:4].astype(int)
        )  # 연도 추출
        reports = reports[
            (reports["연도"] == year)
            & (reports["report_nm"].str.contains("사업보고서"))
        ]

        if reports.empty:
            print(f"⚠ {year}년 {corp_code} 사업보고서 없음")
            continue

        # 📌 가장 오래된 사업보고서의 접수번호 추출
        oldest_rcept_no = reports.iloc[-1]["rcept_no"]

        # 📌 해당 사업보고서에서 주요주주 보유 지분율 가져오기
        stock_data = dart.major_shareholders(corp_code, oldest_rcept_no)

        if stock_data is None or stock_data.empty:
            print(f"⚠ {year}년 {corp_code} 주요 주주 데이터 없음")
            continue

        # 📌 주요 주주의 보유 지분율(`stkrt`) 합산
        stock_data["지분율"] = stock_data["stkrt"].astype(float)
        internal_shareholding = round(stock_data["지분율"].sum(), 2)

        # 📌 데이터 저장
        results.append(
            {
                "종목코드": corp_code,
                "연도": year,
                "내부지분율(%)": internal_shareholding,
            }
        )

    except Exception as e:
        print(f"⚠ {year}년 {corp_code} 데이터 처리 중 오류 발생: {str(e)}")

# 📌 최종 데이터프레임 변환
df = pd.DataFrame(results)
print(df)

# 📌 CSV 저장
output_file = "/Users/seojeonghun/Desktop/DBisFree/과거사업보고서_내부지분율.csv"
df.to_csv(output_file, index=False)
print(f"✅ 과거 내부지분율 데이터가 '{output_file}' 파일로 저장되었습니다.")
```

```python
import requests
import pandas as pd
import zipfile
import xml.etree.ElementTree as ET
import os

# 🔹 DART API Key (여기에 본인의 API 키 입력)
DART_API_KEY = "1ec628ca0ac283731a99b879ea1ab65f79897d24"


# 🔹 기업 코드 데이터 다운로드 및 파싱
def get_corp_codes():
    """
    DART에서 제공하는 전체 기업 리스트 XML을 다운로드하여 기업 코드 데이터를 가져오는 함수
    """
    url = f"https://opendart.fss.or.kr/api/corpCode.xml?crtfc_key={DART_API_KEY}"
    response = requests.get(url)

    zip_path = "corpCode.zip"
    xml_path = "corpCode.xml"

    with open(zip_path, "wb") as f:
        f.write(response.content)

    # ZIP 파일 압축 해제
    with zipfile.ZipFile(zip_path, "r") as zip_ref:
        zip_ref.extractall()

    # XML 파일 파싱
    tree = ET.parse(xml_path)
    root = tree.getroot()

    corp_list = []
    for child in root.findall("list"):
        corp_list.append(
            {
                "corp_code": child.find("corp_code").text.strip(),
                "stock_code": child.find("stock_code").text.strip(),
                "corp_name": child.find("corp_name").text.strip(),
            }
        )

    # 불필요한 파일 삭제
    os.remove(zip_path)
    os.remove(xml_path)

    return pd.DataFrame(corp_list)


# 🔹 특정 기업의 내부지분율 조회
def get_insider_ownership(corp_code):
    """
    특정 기업의 내부지분율(최대주주 및 특수관계인 보유지분)을 조회하는 함수
    """
    url = "https://opendart.fss.or.kr/api/majorstock.json"
    params = {
        "crtfc_key": DART_API_KEY,
        "corp_code": corp_code,
        "bsns_year": "2023",  # 조회 연도 (최신 연도 사용)
        "reprt_code": "11011",  # 사업보고서 기준
    }

    response = requests.get(url, params=params)
    data = response.json()

    if data["status"] != "000":
        return None, None

    records = data.get("list", [])
    if not records:
        return None, None

    df = pd.DataFrame(records)
    df["stkqy"] = pd.to_numeric(df["stkqy"], errors="coerce")  # 주식 수
    df["stkrate"] = pd.to_numeric(df["stkrate"], errors="coerce")  # 지분율(%)

    # 전체 내부 지분율 합산
    total_insider_ownership = df["stkrate"].sum()
    total_stkqy = df["stkqy"].sum()

    return total_stkqy, total_insider_ownership


# 🔹 데이터프레임 생성
def create_insider_ownership_df():
    corp_df = get_corp_codes()

    # 상장 기업만 필터링 (종목코드가 없는 경우 제외)
    corp_df = corp_df[corp_df["stock_code"] != ""]

    results = []
    for _, row in corp_df.iterrows():
        corp_code = row["corp_code"]
        stock_code = row["stock_code"]
        corp_name = row["corp_name"]

        total_stkqy, total_insider_ownership = get_insider_ownership(corp_code)

        results.append(
            {
                "기업종목코드": stock_code,
                "기업코드": corp_code,
                "기업명": corp_name,
                "주식수": total_stkqy,
                "내부지분율": total_insider_ownership,
            }
        )

    return pd.DataFrame(results)


# 🔹 실행 및 데이터프레임 확인
df = create_insider_ownership_df()
import ace_tools as tools

tools.display_dataframe_to_user(name="DART 내부지분율 데이터", dataframe=df)
```

```python
import requests
import pandas as pd

# 🔹 DART API Key (본인의 API 키 입력)
DART_API_KEY = "1ec628ca0ac283731a99b879ea1ab65f79897d24"

# 🔹 삼성전자의 기업 코드 (DART의 고유 기업 코드, 기업 종목코드와 다름)
corp_code = "005930"  # 삼성전자 (테스트용)


# 🔹 특정 기업의 최신 사업보고서(11011) 가져오기
def get_latest_report(corp_code):
    url = "https://opendart.fss.or.kr/api/list.json"
    params = {
        "crtfc_key": DART_API_KEY,
        "corp_code": corp_code,
        "bgn_de": "20230101",  # 2023년 기준
        "end_de": "20231231",
        "pblntf_detail_ty": "A001",  # 정기보고서 (사업보고서)
        "page_no": "1",
        "page_count": "10",
    }

    response = requests.get(url, params=params)
    data = response.json()

    if data["status"] != "000":
        print("📌 사업보고서 조회 실패:", data["message"])
        return None

    reports = data.get("list", [])

    for report in reports:
        if report["report_nm"].startswith("사업보고서"):
            return report["rcept_no"]

    return None


# 🔹 '주주에 관한 사항' 데이터 가져오기
def get_shareholder_info(corp_code, rcept_no):
    url = "https://opendart.fss.or.kr/api/shareholder.json"
    params = {"crtfc_key": DART_API_KEY, "corp_code": corp_code, "rcept_no": rcept_no}

    response = requests.get(url, params=params)
    data = response.json()

    if data["status"] != "000":
        print("📌 주주정보 조회 실패:", data["message"])
        return None

    shareholders = data.get("list", [])

    if not shareholders:
        print("📌 주주 정보 없음")
        return None

    df = pd.DataFrame(shareholders)

    # 🔹 필요한 컬럼만 추출
    df = df[["shrnm", "relate", "shrkind", "shrholdr_stock_co", "shrholdr_irdnt"]]

    # 🔹 컬럼명 변경
    df.columns = ["성명", "관계", "주식 종류", "소유 주식수", "지분율"]

    # 🔹 숫자형 변환
    df["소유 주식수"] = pd.to_numeric(df["소유 주식수"], errors="coerce")
    df["지분율"] = pd.to_numeric(df["지분율"], errors="coerce")

    # 🔹 보통주 지분율 합 계산
    common_share_rate_sum = df[df["주식 종류"] == "보통주"]["지분율"].sum()

    # 🔹 보통주 지분율 합 컬럼 추가
    df["보통주 지분율 합"] = common_share_rate_sum

    return df


# 🔹 실행 및 데이터프레임 생성
rcept_no = get_latest_report(corp_code)
if rcept_no:
    shareholder_df = get_shareholder_info(corp_code, rcept_no)
    import ace_tools as tools

    tools.display_dataframe_to_user(
        name="주주 정보 및 보통주 지분율", dataframe=shareholder_df
    )
```

```python
import OpenDartReader
import pandas as pd
import time

API_KEY = "711145f0ffd8e3f2f02ab901274159f09ace6ef3"  # DART API 키 입력
dart = OpenDartReader(api_key=API_KEY)

# 기업 코드 리스트 (사용자가 입력)
corp_codes = ["005930", "000660"]


# 배당 관련 공시 리스트 가져오기 (첫 페이지만 조회)
def get_dividend_list(start_year=2015, end_year=2024, corp_codes=[]):
    data = []
    for year in range(start_year, end_year + 1):
        for corp_code in corp_codes:
            reports = dart.list(
                corp=corp_code, start=f"{year}0101", end=f"{year}1231", kind="A003"
            )

            # 데이터프레임이 비어 있거나 None이면 스킵
            if reports is None or reports.empty:
                continue

            for _, report in reports.iterrows():
                data.append(
                    {
                        "회사명": report["corp_name"],
                        "공시일": report["rcept_dt"],
                        "공시번호": report["rcept_no"],
                        "기업코드": corp_code,
                    }
                )

            print(f"{corp_code} - {year}년 데이터 처리 완료")
            time.sleep(0.3)  # API 호출 간격 조절
    return pd.DataFrame(data)


# 공시 상세 정보에서 배당기준일 추출
def get_dividend_record_date(recept_no):
    try:
        detail = dart.document(recept_no)
        for key, value in detail.items():
            if "배당기준일" in key:
                return value
    except Exception as e:
        print(f"공시번호 {recept_no} 처리 중 오류 발생: {e}")
    return None


# 기업 코드 리스트 입력 후 실행
df = get_dividend_list(corp_codes=corp_codes)
df["배당기준일"] = df["공시번호"].apply(get_dividend_record_date)

# CSV 파일로 저장
df.to_csv(
    "/Users/seojeonghun/Desktop/DBisFree/0223/공시여부_data.csv",
    index=False,
    encoding="utf-8-sig",
)

print("CSV 파일로 저장 완료: 공시여부_data.csv")
```

```python
import OpenDartReader
import pandas as pd
import time

# 🔹 삼성전자 종목코드
corp_code = "005930"

# 🔹 2024년 연결 재무제표 가져오기 (사업보고서 코드: 11011, separate=False 사용)
df = dart.finstate(corp_code, 2024, reprt_code="11011")  # 연결 재무제표

if df is not None:
    # 🔹 2024년 주요 재무 데이터 필터링
    financial_data = df[df["account_nm"].isin(["매출액", "영업이익", "자산총계"])][
        ["account_nm", "thstrm_amount"]
    ]
    financial_data.columns = ["항목", "2024년 금액"]

    print("✅ 2024년 삼성전자 연결 재무제표 데이터:")
    print(financial_data)
else:
    print("❌ 2024년 데이터가 없습니다.")
```

```python
# 중복 데이터 제거

import pandas as pd

# CSV 파일 불러오기
df = pd.read_csv("/Users/seojeonghun/Desktop/DBisFree/0224/구조방정식_제거전.csv")
print("제거 전")
print(df.info())


# 1️⃣ 종목코드와 연도 기준으로 그룹화 후 배당횟수 기준으로 최대값만 남기기
df_cleaned = df.loc[df.groupby(["corp_code", "year"])["dividend_frequency"].idxmax()]

# 2️⃣ 결과 확인
print(df_cleaned.info())
```

```python
# CSV 파일로 저장
output_file = "/Users/seojeonghun/Desktop/DBisFree/0224/구조방정식_중복제거완료.csv"

# 전체 데이터를 CSV 파일로 저장
df_cleaned.to_csv(output_file, encoding="utf-8", index=False)
```
