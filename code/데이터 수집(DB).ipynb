```python
import requests
import pandas as pd
import zipfile
import xml.etree.ElementTree as ET
import os

# üîπ DART API Key (Ïó¨Í∏∞Ïóê Î≥∏Ïù∏Ïùò API ÌÇ§ ÏûÖÎ†•)
DART_API_KEY = "1ec628ca0ac283731a99b879ea1ab65f79897d24"


# üîπ Í∏∞ÏóÖ ÏΩîÎìú Îç∞Ïù¥ÌÑ∞ Îã§Ïö¥Î°úÎìú Î∞è ÌååÏã±
def get_corp_codes():
    """
    DARTÏóêÏÑú Ï†úÍ≥µÌïòÎäî Ï†ÑÏ≤¥ Í∏∞ÏóÖ Î¶¨Ïä§Ìä∏ XMLÏùÑ Îã§Ïö¥Î°úÎìúÌïòÏó¨ Í∏∞ÏóÖ ÏΩîÎìú Îç∞Ïù¥ÌÑ∞Î•º Í∞ÄÏ†∏Ïò§Îäî Ìï®Ïàò
    """
    url = f"https://opendart.fss.or.kr/api/corpCode.xml?crtfc_key={DART_API_KEY}"
    response = requests.get(url)

    zip_path = "corpCode.zip"
    xml_path = "corpCode.xml"

    with open(zip_path, "wb") as f:
        f.write(response.content)

    # ZIP ÌååÏùº ÏïïÏ∂ï Ìï¥Ï†ú
    with zipfile.ZipFile(zip_path, "r") as zip_ref:
        zip_ref.extractall()

    # XML ÌååÏùº ÌååÏã±
    tree = ET.parse(xml_path)
    root = tree.getroot()

    corp_list = []
    for child in root.findall("list"):
        corp_list.append(
            {
                "corp_code": child.find("corp_code").text.strip(),
                "stock_code": child.find("stock_code").text.strip(),
                "corp_name": child.find("corp_name").text.strip(),
            }
        )

    # Î∂àÌïÑÏöîÌïú ÌååÏùº ÏÇ≠Ï†ú
    os.remove(zip_path)
    os.remove(xml_path)

    return pd.DataFrame(corp_list)


# üîπ ÌäπÏ†ï Í∏∞ÏóÖÏùò ÎÇ¥Î∂ÄÏßÄÎ∂ÑÏú® Ï°∞Ìöå
def get_insider_ownership(corp_code):
    """
    ÌäπÏ†ï Í∏∞ÏóÖÏùò ÎÇ¥Î∂ÄÏßÄÎ∂ÑÏú®(ÏµúÎåÄÏ£ºÏ£º Î∞è ÌäπÏàòÍ¥ÄÍ≥ÑÏù∏ Î≥¥Ïú†ÏßÄÎ∂Ñ)ÏùÑ Ï°∞ÌöåÌïòÎäî Ìï®Ïàò
    """
    url = "https://opendart.fss.or.kr/api/majorstock.json"
    params = {
        "crtfc_key": DART_API_KEY,
        "corp_code": corp_code,
        "bsns_year": "2023",  # Ï°∞Ìöå Ïó∞ÎèÑ (ÏµúÏã† Ïó∞ÎèÑ ÏÇ¨Ïö©)
        "reprt_code": "11011",  # ÏÇ¨ÏóÖÎ≥¥Í≥†ÏÑú Í∏∞Ï§Ä
    }

    response = requests.get(url, params=params)
    data = response.json()

    if data["status"] != "000":
        return None, None

    records = data.get("list", [])
    if not records:
        return None, None

    df = pd.DataFrame(records)
    df["stkqy"] = pd.to_numeric(df["stkqy"], errors="coerce")  # Ï£ºÏãù Ïàò
    df["stkrate"] = pd.to_numeric(df["stkrate"], errors="coerce")  # ÏßÄÎ∂ÑÏú®(%)

    # Ï†ÑÏ≤¥ ÎÇ¥Î∂Ä ÏßÄÎ∂ÑÏú® Ìï©ÏÇ∞
    total_insider_ownership = df["stkrate"].sum()
    total_stkqy = df["stkqy"].sum()

    return total_stkqy, total_insider_ownership


# üîπ Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ ÏÉùÏÑ±
def create_insider_ownership_df():
    corp_df = get_corp_codes()

    # ÏÉÅÏû• Í∏∞ÏóÖÎßå ÌïÑÌÑ∞ÎßÅ (Ï¢ÖÎ™©ÏΩîÎìúÍ∞Ä ÏóÜÎäî Í≤ΩÏö∞ Ï†úÏô∏)
    corp_df = corp_df[corp_df["stock_code"] != ""]

    results = []
    for _, row in corp_df.iterrows():
        corp_code = row["corp_code"]
        stock_code = row["stock_code"]
        corp_name = row["corp_name"]

        total_stkqy, total_insider_ownership = get_insider_ownership(corp_code)

        results.append(
            {
                "Í∏∞ÏóÖÏ¢ÖÎ™©ÏΩîÎìú": stock_code,
                "Í∏∞ÏóÖÏΩîÎìú": corp_code,
                "Í∏∞ÏóÖÎ™Ö": corp_name,
                "Ï£ºÏãùÏàò": total_stkqy,
                "ÎÇ¥Î∂ÄÏßÄÎ∂ÑÏú®": total_insider_ownership,
            }
        )

    return pd.DataFrame(results)


# üîπ Ïã§Ìñâ Î∞è Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ ÌôïÏù∏
df = create_insider_ownership_df()
import ace_tools as tools

tools.display_dataframe_to_user(name="DART ÎÇ¥Î∂ÄÏßÄÎ∂ÑÏú® Îç∞Ïù¥ÌÑ∞", dataframe=df)
```

```python
import OpenDartReader
import pandas as pd
import time

API_KEY = "711145f0ffd8e3f2f02ab901274159f09ace6ef3"  # DART API ÌÇ§ ÏûÖÎ†•
dart = OpenDartReader(api_key=API_KEY)

# Í∏∞ÏóÖ ÏΩîÎìú Î¶¨Ïä§Ìä∏ (ÏÇ¨Ïö©ÏûêÍ∞Ä ÏûÖÎ†•)
corp_codes = ["005930", "000660"]


# Î∞∞Îãπ Í¥ÄÎ†® Í≥µÏãú Î¶¨Ïä§Ìä∏ Í∞ÄÏ†∏Ïò§Í∏∞ (Ï≤´ ÌéòÏù¥ÏßÄÎßå Ï°∞Ìöå)
def get_dividend_list(start_year=2015, end_year=2024, corp_codes=[]):
    data = []
    for year in range(start_year, end_year + 1):
        for corp_code in corp_codes:
            reports = dart.list(
                corp=corp_code, start=f"{year}0101", end=f"{year}1231", kind="A003"
            )

            # Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏù¥ ÎπÑÏñ¥ ÏûàÍ±∞ÎÇò NoneÏù¥Î©¥ Ïä§ÌÇµ
            if reports is None or reports.empty:
                continue

            for _, report in reports.iterrows():
                data.append(
                    {
                        "ÌöåÏÇ¨Î™Ö": report["corp_name"],
                        "Í≥µÏãúÏùº": report["rcept_dt"],
                        "Í≥µÏãúÎ≤àÌò∏": report["rcept_no"],
                        "Í∏∞ÏóÖÏΩîÎìú": corp_code,
                    }
                )

            print(f"{corp_code} - {year}ÎÖÑ Îç∞Ïù¥ÌÑ∞ Ï≤òÎ¶¨ ÏôÑÎ£å")
            time.sleep(0.3)  # API Ìò∏Ï∂ú Í∞ÑÍ≤© Ï°∞Ï†à
    return pd.DataFrame(data)


# Í≥µÏãú ÏÉÅÏÑ∏ Ï†ïÎ≥¥ÏóêÏÑú Î∞∞ÎãπÍ∏∞Ï§ÄÏùº Ï∂îÏ∂ú
def get_dividend_record_date(recept_no):
    try:
        detail = dart.document(recept_no)
        for key, value in detail.items():
            if "Î∞∞ÎãπÍ∏∞Ï§ÄÏùº" in key:
                return value
    except Exception as e:
        print(f"Í≥µÏãúÎ≤àÌò∏ {recept_no} Ï≤òÎ¶¨ Ï§ë Ïò§Î•ò Î∞úÏÉù: {e}")
    return None


# Í∏∞ÏóÖ ÏΩîÎìú Î¶¨Ïä§Ìä∏ ÏûÖÎ†• ÌõÑ Ïã§Ìñâ
df = get_dividend_list(corp_codes=corp_codes)
df["Î∞∞ÎãπÍ∏∞Ï§ÄÏùº"] = df["Í≥µÏãúÎ≤àÌò∏"].apply(get_dividend_record_date)

# CSV ÌååÏùºÎ°ú Ï†ÄÏû•
df.to_csv(
    "/Users/seojeonghun/Desktop/DBisFree/0223/Í≥µÏãúÏó¨Î∂Ä_data.csv",
    index=False,
    encoding="utf-8-sig",
)

print("CSV ÌååÏùºÎ°ú Ï†ÄÏû• ÏôÑÎ£å: Í≥µÏãúÏó¨Î∂Ä_data.csv")
```
